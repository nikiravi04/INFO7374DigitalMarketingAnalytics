# Use MAX( viewers ) table to filter out duplicate rows (from genre)

```python
query = '''SELECT user_id, user_name,count(distinct(game_id)) as num_games, MAX(viewer_count) as max_viewers
FROM stream_data 
GROUP BY user_id, user_name
HAVING count(distinct(game_id))>1
ORDER BY num_games DESC, max_viewers DESC;'''
df_max_viewers = pd_sql.read_sql(query, connection)
```

## Getting the data for surprise recommender in the format to be funneled into the Surprise notebook

```python
query = "SELECT Count(*) FROM general_game_info;"

pd_sql.read_sql(query, connection)
```

```python
query = "SELECT * FROM general_game_info LIMIT 5;"

pd_sql.read_sql(query, connection)
```

Surprise needs numeric ratings between 1-5 for every streamer and game they've played. When we tried to make a _user x game_ matrix (something in the ballpark of 55,000 users by 162 games), this matrix was too big to operate on locally.


To give ratings, we look at the **MAX** viewers the streamer has had a in one stream session in a particular game across all time periods

This is done using this query:

```python
query = '''SELECT user_name, user_id,game_id,game_name, game_genres, language, started_at, viewer_count, max(viewer_count)
FROM general_game_info
GROUP BY user_name, user_id,game_id,game_name, game_genres, language, started_at, viewer_count
ORDER BY user_name DESC, max DESC'''

surprise_grid = pd_sql.read_sql(query, connection)
```

```python
surprise_grid.info()
```

```python
surprise_grid.to_csv('final_game_user_grid.csv', index = False)
```

```python
query = '''SELECT user_name, game_name, max(viewer_count)
FROM general_game_info
GROUP BY user_name, game_name
ORDER BY user_name DESC, max DESC'''

surprise_grid_minimal = pd_sql.read_sql(query, connection)
```
